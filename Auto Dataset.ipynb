```python
filename = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df=pd.read_csv(filename,names=headers)
df.head()
```

# Convert "?" to NaN

```python
auto=df.replace('?',np.nan)
auto1=auto.isnull()

```

# drop data

## drop the whole row price where price is null value

```python
auto.dropna(subset=['price'],inplace=True) 
```

# fill the columns with mean values

```python
auto.info()
```

```python
mean=auto[['normalized-losses','bore','stroke','peak-rpm','horsepower']].astype(float).mean()
mean
```

```python
auto[['normalized-losses','bore','stroke','peak-rpm','horsepower']].replace(to_replace=np.nan,value=mean)
auto
```

```python
auto['num-of-doors'].value_counts()
```

```python
auto['num-of-doors'].replace('two',2,inplace=True)
auto['num-of-doors'].replace('four',4,inplace=True)
auto['num-of-doors'].fillna(auto['num-of-doors'].astype(float).mode(),inplace=True)
auto['num-of-doors'].replace(2.0,4.0,inplace=True)
auto.head()
```
